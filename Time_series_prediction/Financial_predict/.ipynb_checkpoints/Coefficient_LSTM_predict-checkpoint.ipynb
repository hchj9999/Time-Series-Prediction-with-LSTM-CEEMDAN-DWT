```python
# 取一段時間的金融資料，傳給several_CEEMDAN()
def take_latest_several_data(end_date,timerange):
    
    end_date = datetime.strptime(end_date, '%Y-%m-%d').date()
    start_date = end_date - timedelta(days=timerange)
    
    financial_new_data = yf.download("^TWII", start=start_date, end=end_date)
    financial_new_data = financial_new_data.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)
    financial_new_data = financial_new_data.sort_index(ascending=False)
    financial_new_data['Close'] = financial_new_data['Close'].replace('', np.nan)
    financial_new_data['Close'] = financial_new_data['Close'].astype(float) 
    
    financial_new_data_df = financial_new_data
    financial_new_data = financial_new_data['Close'].to_list()
    
    return financial_new_data_df,financial_new_data
```

```python
# 利用一段時間的金融資料做少量ceemdan分解，回傳分解結果(幾個coefficient對應幾個資料點)
def several_SWT(end_date,assume_shortest_decompose_length,timerange,coefficient_num):
    financial_new_data_df,financial_new_data = take_latest_several_data(end_date,timerange)
#     reversed_financial_new_data = financial_new_data[::-1] 
    
    start = time.time()
    
    new_coefficients = pywt.swt(np.array(financial_new_data[:assume_shortest_decompose_length]), 'db10', level=coefficient_num)
    
#     for i in range(len(financial_new_data)):
#         ceemdan.ceemdan(np.array(financial_new_data[:assume_shortest_decompose_length+i]).ravel(),max_imf=imf_num)
#         new_imfs, new_res = ceemdan.get_imfs_and_residue()
#         if len(new_imfs) == imf_num+1:
#             break

    end = time.time()
    print("date:",end_date,"分解執行時間：%f 秒" % (end - start),"Coefficient數量：",len(new_coefficients),"資料點數量：",len(new_coefficients[0][0]),"真實資料：",financial_new_data_df[:assume_shortest_decompose_length])
#     several_CEEMDAN_plot(new_imfs)
    
#     ceemdan_decompose_res(new_imfs)
    return new_coefficients
```

```python
# 用分解後的結果更新資料庫中對應的coefficient資料表
def New_coefficient_to_Database(end_date,assume_shortest_decompose_length,timerange,coefficient_num): 
    
    data_to_update = several_SWT(end_date,assume_shortest_decompose_length,timerange,coefficient_num)
    financial_new_data_df,financial_new_data = take_latest_several_data(end_date,timerange)
#     print(data_to_update)
    c_number = 0
    for i in range(len(data_to_update)):
        for j in range(len(data_to_update[i])):
            c_number = c_number+1
            coefficient_df = pd.DataFrame(data_to_update[i][j])
            coefficient_df.set_index(financial_new_data_df[:len(data_to_update[i][j])].index, inplace=True)
            coefficient_df = coefficient_df.rename(columns={0: 'Close_^TWII_coefficient'})
            coefficient_df = coefficient_df.reset_index()
            connection = pymysql.connect(host = '140.116.39.172' ,  user = 'KDD' ,  passwd = "kdd172", db='financial_day')
            cursor = connection.cursor()
            data = coefficient_df.values.tolist()
            insert_query = '''
            REPLACE INTO coefficient{} (`Date`, `coefficient{}`)
            VALUES (%s, %s)
            '''.format(c_number,c_number)
            cursor.executemany(insert_query, data)
            connection.commit()
            connection.close()
    print("更新資料表")
    return data_to_update
```

```python
def createSlidingWindow(dataset,windowSize):
    x_train = []
    y_train = []
    for i in range(len(dataset)-windowSize):
        x_train.append(dataset[i:i+windowSize])
        y_train.append(dataset[i+windowSize])
    return np.array(x_train), np.array(y_train)
```

```python
# 參數轉成監督式參數 [B]
# The function is defined with default parameters so that if you call it with just your data, it will construct a DataFrame with t-1 
# as X and t as y

def series_to_supervised(data, n_in = 1, n_out = 1, dropnan = True):
    n_vars = 1 if type(data) is list else data.shape[1]
    
    df = pd.DataFrame(data)
    df_columns_name = df.columns.values
    cols, names = list(), list()
    
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('%s(t-%d)' % (df_columns_name[j], i)) for j in range(n_vars)]
        
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('%s(t)' % (df_columns_name[j])) for j in range(n_vars)]
        else:
            names += [('%s(t+%d)' % (df_columns_name[j], i)) for j in range(n_vars)]
            
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
        
    return agg
# 參數轉成監督式參數 [E]
```

```python
def predict_multistep(model, input_data, n_steps_out):
    predictions = []
    current_input = np.expand_dims(input_data, axis=0)

    for _ in range(n_steps_out):
        prediction = model.predict(current_input)[0]
        predictions.append(prediction)
        current_input = np.append(current_input[:, 1:], np.expand_dims(prediction, axis=1), axis=1)
 
    return np.array(predictions)
```

```python
# 取日期範圍
def datelist(start_date,end_date):
    date_list = []
    start_date = datetime.strptime(start_date, '%Y-%m-%d')  # 起始日期
    end_date = datetime.strptime(end_date, '%Y-%m-%d')    # 结束日期
    current_date = start_date
    while current_date <= end_date:
        date_list.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)
    return date_list
```

```python
def predict_date_range(date_range_list,assume_shortest_decompose_length,timerange,coefficient_num):
    predict_list = []
    execute_time_list = []
    all_time_start = time.time()
    for i in range(len(date_range_list)):
        start = time.time()
        data_to_update = New_coefficient_to_Database(date_range_list[i],assume_shortest_decompose_length,timerange,coefficient_num)
    

        coefficient1_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient1', 'vanilla', 1, './model/SWT_simple-LSTM-model-result-1-Copy1.h5')
        coefficient2_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient2', 'vanilla', 2, './model/SWT_simple-LSTM-model-result-2-Copy1.h5')
        coefficient3_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient3', 'vanilla', 3, './model/SWT_simple-LSTM-model-result-3-Copy1.h5')
        coefficient4_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient4', 'vanilla', 4, './model/SWT_simple-LSTM-model-result-4-Copy1.h5')
        coefficient5_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient5', 'vanilla', 5, './model/SWT_simple-LSTM-model-result-5-Copy1.h5')
        coefficient6_result = predict_financial('{}'.format(date_range_list[i]), 'coefficient6', 'vanilla', 6, './model/SWT_simple-LSTM-model-result-6-Copy1.h5')
    
        result_arr1 = []
        result_arr2 = []
        result_arr3 = []
        result_arr4 = []
        result_arr5 = []
        result_arr6 = []
        
        result_arr1.append(coefficient1_result)
        result_arr2.append(coefficient2_result)
        result_arr3.append(coefficient3_result)
        result_arr4.append(coefficient4_result)
        result_arr5.append(coefficient5_result)
        result_arr6.append(coefficient6_result)
        
        for i in range(assume_shortest_decompose_length-1):
            result_arr1.append(data_to_update[0][0][i])
            result_arr2.append(data_to_update[0][1][i])
            result_arr3.append(data_to_update[1][0][i])
            result_arr4.append(data_to_update[1][1][i])
            result_arr5.append(data_to_update[2][0][i])
            result_arr6.append(data_to_update[2][1][i])
        
        result_arr1 = np.array(result_arr1)
        result_arr2 = np.array(result_arr2)
        result_arr3 = np.array(result_arr3)
        result_arr4 = np.array(result_arr4)
        result_arr5 = np.array(result_arr5)
        result_arr6 = np.array(result_arr6)
        
        coefficient_predict_reconstruct = SWT_reconstruct(result_arr1,result_arr2,result_arr3,result_arr4,result_arr5,result_arr6)
        predict_list.append(coefficient_predict_reconstruct[0])
        end = time.time()
        execute_time_list.append(end - start)
        print("日期：",date_range_list[i],"(加上前七天)對應的預測值",coefficient_predict_reconstruct,"一天的分解+預測執行時間：", (end - start))
    all_time_end = time.time()
    print("全部執行時間：%f 秒" % (all_time_end - all_time_start))
    return predict_list, execute_time_list
```

# 比較實際值與預測結果

```python
# 取真實金融資料的dataframe
def get_true_value(start_day,end_day):
    conn  =  pymysql.connect ( host = '140.116.39.172' ,  user = 'KDD' ,  passwd = "kdd172" ,  db = 'financial_day' ) 
    cur  =  conn.cursor() 
    select_sql = '''SELECT `Close_^TWII`, `Date`
                    FROM `financial_data` 
                    WHERE `Date` >= '{}' AND `Date` <= '{}' '''.format(start_day, end_day)
    result_object = cur.execute(select_sql)
    results_values_list = cur.fetchall()
    result_key_list = [i[0] for i in cur.description]
    true_day_value = pd.DataFrame(results_values_list)
    true_day_value.columns = result_key_list
    true_day_value = true_day_value.set_index('Date')
    cur.close () 
    conn.close()

    true_day_value['Close_^TWII'] = true_day_value['Close_^TWII'].replace('', np.nan)
    true_day_value['Close_^TWII'] = true_day_value['Close_^TWII'].astype(float)  
    return true_day_value
    # datelist = true_day_value.index
```

```python
# 將真實資料與預測值合併成一個dataframe
def concat_true_and_predict_value(start_day, end_day, predict_list,execute_list):
    true_day_value = get_true_value(start_day,end_day)
    predict_datelist = datelist(start_day,end_day)
    predict_df = pd.DataFrame({'Date': predict_datelist, 'predict': predict_list})
    execute_time_df = pd.DataFrame({'Date': predict_datelist, 'execute_time': execute_list})
    predict_df = predict_df.set_index('Date')
    execute_time_df = execute_time_df.set_index('Date')
    merged_df = pd.concat([true_day_value, predict_df], axis=1)
    merged_df = pd.concat([merged_df, execute_time_df], axis=1)
    merged_df = merged_df.dropna()
    return merged_df
```
