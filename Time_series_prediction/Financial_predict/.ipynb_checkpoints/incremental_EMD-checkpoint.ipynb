```python
begin_date='2023-01-28'
end_date='2023-05-28'
```

```python
TWII = yf.download('^TWII',start=begin_date,end=end_date)
TWII_Close = TWII['Close']
TWII_Close = pd.DataFrame(TWII_Close)
len(TWII_Close)
```

```python
TWII_slide = yf.download('^TWII',start='2023-01-31',end='2023-05-30')
TWII_Close_slide = TWII_slide['Close']
TWII_Close_slide = pd.DataFrame(TWII_Close_slide)
len(TWII_Close_slide)
```

```python
TWII_Close_arr=np.array(TWII_Close)
decomposer = EMD(TWII_Close_arr,n_imfs=4)
TWII_Close_arr_imfs = decomposer.decompose()
fig = pyplot.gcf()
fig.set_size_inches(18.5, 10.5)


TWII_Close_arr2=np.array(TWII_Close_slide)
decomposer2 = EMD(TWII_Close_arr2,n_imfs=4)
TWII_Close_arr_imfs2 = decomposer2.decompose()
fig = pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

plot_imfs(TWII_Close_arr,TWII_Close_arr_imfs)
plot_imfs(TWII_Close_arr2,TWII_Close_arr_imfs2)

# plt.plot(TWII_Close_arr_imfs[0])
# plt.plot(TWII_Close_arr_imfs2[0],color='y')

# plt.plot(TWII_Close_arr_imfs[1])
# plt.plot(TWII_Close_arr_imfs2[1],color='y')

# plt.plot(TWII_Close_arr_imfs[2])
# plt.plot(TWII_Close_arr_imfs2[2],color='y')
```
