```python
# 參數轉成監督式參數 [B]
# The function is defined with default parameters so that if you call it with just your data, it will construct a DataFrame with t-1 
# as X and t as y

def series_to_supervised(data, n_in = 1, n_out = 1, dropnan = True):
    n_vars = 1 if type(data) is list else data.shape[1]
    
    df = pd.DataFrame(data)
    df_columns_name = df.columns.values
    cols, names = list(), list()
    
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('%s(t-%d)' % (df_columns_name[j], i)) for j in range(n_vars)]
        
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('%s(t)' % (df_columns_name[j])) for j in range(n_vars)]
        else:
            names += [('%s(t+%d)' % (df_columns_name[j], i)) for j in range(n_vars)]
            
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
        
    return agg
# 參數轉成監督式參數 [E]
```

```python
# 所有使用到的變量
train_features_df = pd.concat([TWII_train_recolumn, dji_train_recolumn, china_train_recolumn], axis=1)
train_features_df = train_features_df.drop(train_features_df[train_features_df['twii_Open'].isna()].index)
# train_features_df
train_features_df = train_features_df.interpolate(method='time')
train_features_df = train_features_df.tail(1964)
train_features_redf = series_to_supervised(train_features_df)
# features_df[features_df.isna().any(axis=1)]
train_features_redf
```

```python
# 所有使用到的變量
predict_features_df = pd.concat([TWII_predict_recolumn, dji_predict_recolumn, china_predict_recolumn], axis=1)
predict_features_df = predict_features_df.drop(predict_features_df[predict_features_df['twii_Open'].isna()].index)
# train_features_df
predict_features_df = predict_features_df.interpolate(method='time')
predict_features_df = predict_features_df.tail(489)
predict_features_redf = series_to_supervised(predict_features_df)
# features_df[features_df.isna().any(axis=1)]
predict_features_redf
```

```python
select_column = ['twii_Open(t)','twii_High(t)','twii_Low(t)','dji_Open(t)','dji_High(t)','dji_Low(t)','dji_Close(t)','china_Open(t)',
                'china_High(t)','china_Low(t)','china_Close(t)','twii_Close(t-1)']
# values = features_df.values
all_train_features = train_features_redf.loc[train_begin_date:train_end_date, select_column]
all_train_labels = train_features_redf.loc[train_begin_date:train_end_date, 'twii_Close(t)']

all_predict_features = predict_features_redf.loc[predict_begin_date:predict_end_date, select_column]
all_predict_labels = predict_features_redf.loc[predict_begin_date:predict_end_date, 'twii_Close(t)']

# values = values.astype('float32')
scaler = MinMaxScaler(feature_range = (0,1))
#train的features,labels
all_train_features_scaled = scaler.fit_transform(all_train_features.values)
all_train_features_scaled = pd.DataFrame(all_train_features_scaled, columns = select_column)

all_train_labels_scaled = scaler.fit_transform(all_train_labels.values.reshape(-1,1))
all_train_labels_scaled = pd.DataFrame(all_train_labels_scaled, columns = ['twii_Close(t)'])

#predict的features,labels
all_predict_features_scaled = scaler.fit_transform(all_predict_features.values)
all_predict_features_scaled = pd.DataFrame(all_predict_features_scaled)

all_predict_labels_scaled = scaler.fit_transform(all_predict_labels.values.reshape(-1,1))
all_predict_labels_scaled = pd.DataFrame(all_predict_labels_scaled, columns = ['twii_Close(t)'])
# all_train_features_scaled
```

# 設定sliding window

```python
# def createSlidingWindow(dataset,windowSize):
#     x_train = []
#     y_train = []
#     for i in range(len(dataset)-windowSize):
#         x_train.append(dataset[i:i+windowSize])
#         y_train.append(dataset[i+windowSize])
#     return np.array(x_train), np.array(y_train)
```

```python
def createSlidingWindow(train_features,train_labels,timestamp=1):
    X_train,y_train = [], []
    for i in range(train_labels.shape[0] - (timestamp - 1)):
        X_train.append(train_features.iloc[i:i+timestamp].values)
        y_train.append(train_labels.iloc[i + (timestamp - 1)])
    X_train, y_train = np.array(X_train), np.array(y_train).reshape(-1,1)
    return X_train,y_train
```

# 選定幾個時間點為一個window

```python
#拆分成訓練集與測試集
timestamp = 30
x_train,y_train = createSlidingWindow(all_train_features_scaled,all_train_labels_scaled,timestamp)
x_test,y_test = createSlidingWindow(all_predict_features_scaled,all_predict_labels_scaled,timestamp)
# x_train,y_train = createSlidingWindow(all_train_features,all_train_labels,timestamp)
# x_test,y_test = createSlidingWindow(all_predict_features,all_predict_labels,timestamp)
```

# 創建model

```python
#Build LSTM model
model=Sequential()
#Add first layer to model
model.add(LSTM(32, return_sequences=True, input_shape=(timestamp,x_train.shape[2])))
#Add second layer to model
# model.add(LSTM(4, return_sequences=False))
# #Add Dense Layer to model with 25 neurons
# model.add(Dense(4))
#Add Dense Layer to model with 1 neuron
# model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

# 訓練model

```python
start = time.time()
model_history = model.fit(x_train,y_train,batch_size=60,epochs=100)
end = time.time()
print("執行時間：%f 秒" % (end - start))
```

```python
test_predict = model.predict(x_test)
# test_predict_supervised = series_to_supervised(test_predict,1,1)
predict_Close = scaler.inverse_transform(test_predict.reshape(-1,30))
y_test_true = scaler.inverse_transform(y_test.reshape(-1,1))
predict_Close
```

```python
plt.plot(model_history.history['loss'])
```

```python
plt.plot(model_history.history['accuracy'])
```

```python
k = y_test_true.flatten()
r = predict_Close[:,-1].flatten()
```

```python
LSTM_R2 = r2_score(y_test_true,predict_Close[:,-1])
LSTM_MSE = mean_squared_error(y_test_true,predict_Close[:,-1])
LSTM_MAE = mean_absolute_error(y_test_true,predict_Close[:,-1])
LSTM_MRE = np.mean(np.abs((k - r) /k)) * 100
print("LSTM_R2=",LSTM_R2)
print("LSTM_MSE=",LSTM_MSE)
print("LSTM_MAE=",LSTM_MAE)
print("LSTM_RMSE=",sqrt(LSTM_MSE))
print("LSTM_MRE=",LSTM_MRE)
plt.figure(figsize=(10 ,5))
plt.title("window_size={}".format(timestamp))
plt.plot(y_test_true)
plt.plot(predict_Close[:, -1],c='g')
plt.legend(['data', 'LSTM_prediction'], loc='upper right')
plt.show()
```
