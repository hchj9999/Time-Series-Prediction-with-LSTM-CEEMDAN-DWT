```python
random_numbers = []
for i in range(20):
    random_numbers.append(random.randint(1, 100))
print(random_numbers)
```

```python
random_numbers_slide = []
for i in range(len(random_numbers)-1):
    random_numbers_slide.append(random_numbers[i+1])
random_numbers_slide.append(random_numbers[0])
random_numbers_slide
```

```python
random_numbers_arr=np.array(random_numbers)
decomposer = EMD(random_numbers_arr,n_imfs=1)
imfs = decomposer.decompose()
fig = pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
plot_imfs(random_numbers_arr,imfs)
```
